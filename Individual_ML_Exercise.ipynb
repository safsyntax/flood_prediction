***Splitting data then creating train and test sets***

```python
data = pd.read_csv('/content/drive/MyDrive/Machine Learning - Individual Project Fall 24/data.csv')

#Target Variable
y = data['FloodProbability']

# Features
features = ['MonsoonIntensity', 'TopographyDrainage', 'RiverManagement','Deforestation','Urbanization','ClimateChange', 'DamsQuality','Siltation','AgriculturalPractices','Encroachments','IneffectiveDisasterPreparedness','DrainageSystems','CoastalVulnerability','Landslides','Watersheds','DeterioratingInfrastructure','PopulationScore','WetlandLoss','InadequatePlanning','PoliticalFactors']

X = data[features]

#Importing the train_test_split fn.
from sklearn.model_selection import train_test_split

#Splitting my data to have train and test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=79359517)
```

***Performing Recursive Feature Elimination***

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import make_friedman1
from sklearn.feature_selection import RFE
from sklearn.svm import SVR

dt_regressor = DecisionTreeRegressor(random_state=79359517)

X, y = make_friedman1(n_samples=50000, n_features=20, random_state=79359517)
dt_regressor=SVR(kernel="linear")

selector=RFE(dt_regressor, n_features_to_select=10, step=1)
selector=selector.fit(X_train, y_train)

selected_features=X_train.columns[selector.support_]
print("RFE Selected Features: ", selected_features)

selector.ranking_
```


***Adjustments for selected features only***

```python
X_train_selected=X_train[selected_features]
X_test_selected=X_test[selected_features]
```

***Creating and training the Decision Tree Regressor***

```python
from sklearn.tree import DecisionTreeRegressor

dt_regressor = DecisionTreeRegressor(random_state=79359517)
dt_regressor = dt_regressor.fit(X_train_selected,y_train)
```

***Making Predictions***

```python
y_pred=dt_regressor.predict(X_test_selected)
```

***Evalutating the model***

```python
from sklearn import metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error: ", mae)

r_sq = r2_score(y_test, y_pred)
print("R-squared: ", r_sq)
```

***Switching to Linear Model since DT Regressor is giving a negative r-sq value***




```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X_train_selected, y_train)
```

***Predicting FloodProbability (target variable) for the test set***

```python
y_pred = linear_model.predict(X_test_selected)
```

***Evaluating this Model***

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error: ", mae)

r_sq = r2_score(y_test, y_pred)
print("R-squared: ", r_sq)
```

```python
#Some code to check the value of the adjusted r_sq
p=10
n=50000
adjusted_r2 = 1 - (1 - (r_sq)) * (n - 1) / (n - p - 1)
adjusted_r2
```
